```python
df = df.drop_duplicates().reset_index(drop=True)
df.shape
```

```python
df['company'] = df['company'].fillna(0)
```

```python
df['agent'] = df['agent'].fillna(0)
```

```python
df['country'] = df['country'].fillna('Unknown')
```

```python
df['children'] = df['children'].fillna(0)
```

```python
df.isnull().sum()
```

Convert Date Columns and combine them into one proper date column:

```python
df['arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month'].astype(str) + '-' +
    df['arrival_date_day_of_month'].astype(str),
    errors='coerce'
)
```

Convert children, babies, adults to integer

```python
df['children'] = df['children'].astype(int)
df['babies'] = df['babies'].astype(int)
df['adults'] = df['adults'].astype(int)
```

Ensuring categorical columns are converted properly

```python
cat_cols = [
    'hotel', 'meal', 'market_segment', 'distribution_channel',
    'reserved_room_type', 'assigned_room_type',
    'deposit_type', 'customer_type', 'country', 'reservation_status'
]

for col in cat_cols:
    df[col] = df[col].astype('category')
```

Converting agent & company to integer (after filling missing)

```python
df['agent'] = df['agent'].astype(int)
df['company'] = df['company'].astype(int)
```

New Columns

```python
# 1. Total stay duration

df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
```

```python
#2 Total guests

df['total_guests'] = df['adults'] + df['children'] + df['babies']
```

```python
#3 Month name (better for plots)

df['arrival_month_name'] = df['arrival_date'].dt.month_name()
```

```python
#4 Day of week

df['arrival_day_of_week'] = df['arrival_date'].dt.day_name()
```

**Visualization 1 - Number of Bookings by Hotel Type**

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='hotel', palette='Set2')
plt.title('Number of Bookings by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Booking Count')
plt.show()
```

**Visualization 2 - Monthly Booking Trend**

```python
plt.figure(figsize=(12,5))
order = ['January','February','March','April','May','June',
         'July','August','September','October','November','December']

sns.countplot(data=df, x='arrival_month_name', order=order)
plt.title('Monthly Booking Trend')
plt.xticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Bookings')
plt.show()
```

**Visualization 3 - Cancellation Rate by Hotel Type**

```python
cancel_rate = df.groupby('hotel')['is_canceled'].mean().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(data=cancel_rate, x='hotel', y='is_canceled', palette='coolwarm')
plt.title('Cancellation Rate by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Cancellation Rate')
plt.show()
```

#**Key Insights From Exploratory Data Analysis**

**1. Bookings by Hotel Type**

City Hotels receive significantly more bookings than Resort Hotels.

This signals higher demand for short, business-oriented stays in urban locations.

**2. Monthly Booking Trend**

Bookings peak strongly in June–August, driven by summer holidays.

Demand dips in November–January, confirming clear seasonality.

**3. Cancellation Rate by Hotel Type**

City Hotels show higher cancellation rates than Resorts.

Likely due to business travelers’ flexible schedules and last-minute changes.

**4. Lead Time Distribution**

Most bookings occur within 0–30 days of arrival.

A long right-tail indicates occasional early planners, but majority are short-window decisions.

**5. Cancellation Rate by Distribution Channel**

OTA and TA/TO channels have the highest cancellation ratios.

Direct bookings are more stable with far fewer cancellations.

**6. Top 10 Countries by Bookings**

Bookings are heavily concentrated in Portugal, UK, France, Spain, and Germany.

Booking.com has a strong Western European customer base, with Portugal as the dominant contributor.

**7. Total Nights Distribution**

Most stays fall within 1–7 nights, confirming preference for short trips and business travel.

Long stays (10+ nights) are uncommon.

**8. Booking Patterns by Day of Week**

Monday has the highest number of arrivals, indicating strong business travel.

Friday and Saturday follow, driven by leisure and weekend trips.

Tuesday and Sunday record the lowest arrivals.

# **SECTION 2 - ASSUMPTION VALIDATION**

**Assumption Validation**
**1. City hotels are more prone to cancellations.**

Validated – TRUE
City Hotels show a noticeably higher cancellation rate compared to Resort Hotels, supporting the expectation that business-heavy bookings tend to cancel more frequently.

**2. Bookings are highest on weekends.**

Partially True
Friday and Saturday record strong booking volumes.
However, Monday shows the highest overall bookings, driven by business travel.
Weekend demand is high, but not the peak.

**3. Booking trends show clear seasonality.**

Validated – TRUE
Bookings peak in June–August (summer season) and drop significantly between October–January, confirming strong seasonal patterns.

**4. Certain distribution channels have higher cancellation rates.**

Validated – TRUE
OTA and TA/TO channels exhibit the highest cancellation ratios.
Direct bookings show the lowest cancellation rate, making them more reliable.

**5. Guests typically book short stays.**

Validated – TRUE
Total nights distribution shows the majority of stays fall within 1–7 nights, confirming preference for short-duration travel.

**6. Lead time affects booking behavior.**

Validated – TRUE
Most bookings are made within 0–30 days of arrival, with long lead times being uncommon.

**7. A few countries dominate booking volume.**

Validated – TRUE
Portugal, UK, France, Spain, and Germany account for the majority of bookings, indicating a strong Western European customer base.
